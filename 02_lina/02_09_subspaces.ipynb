# 2.9 Subspaces

> **Vector Space.** A real-valued vector space $V = (\mathcal{V}, +, \cdot)$ is a set $\mathcal{V}$ with two operations:
> $$\begin{align}
    +: & \mathcal{V} \times \mathcal{V} \mapsto \mathcal{V}\\
    \cdot: & \mathbb{R} \times \mathcal{V} \mapsto \mathcal{V}
    \end{align}$$
> where:
> 1. $(\mathcal{V}, +)$ is an Abelian Group
> 2. Distributivity
> 3. Associativity (outer operation $\cdot$):
> 4. Neutral element (outer operation $\cdot$):

> **Vector Subspace.** Let $V = (\mathcal{V}, +, \cdot)$ be a vector space and $\mathcal{U} \subseteq \mathcal{V}$, $\mathcal{U} \neq \emptyset$. Then $U = (\mathcal{U}, +, \cdot)$ is called vector subspace of $V$, if $U$ is a vector space with the vector space operations $+$ and $\cdot$ restricted to $\mathcal{U} \times \mathcal{U}$ and $\mathbb{R} \times \mathcal{U}$.

> Which of the following sets are subspaces of $\mathbb{R}^3$?

Approach: Test if it is a vector space assuming the default $+$ and $\cdot$ operations

> (a) $A = \{ (\lambda, \lambda + \mu^3, \lambda - \mu^3), \lambda, \mu \in \mathbb{R} \}$

$A$ is a subspace of $\mathbb{R}^3$:

1. $(A, +)$ is an Abelian Group:
    1. Closure:
    $$
        \begin{align}
            \begin{pmatrix}
                \lambda_1\\
                \lambda_1 + \mu_1^3\\
                \lambda_1 - \mu_1^3
            \end{pmatrix}+
            \begin{pmatrix}
                \lambda_2\\
                \lambda_2 + \mu_2^3\\
                \lambda_2 - \mu_2^3
            \end{pmatrix} &= 
            \begin{pmatrix}
                \lambda_1 + \lambda_2\\
                (\lambda_1 + \lambda_2) + (\mu_1^3 + \mu_2^3)\\
                (\lambda_1 + \lambda_2) - (\mu_1^3 + \mu_2^3)\\
            \end{pmatrix}\\
            &= 
            \begin{pmatrix}
                \lambda_3\\
                \lambda_3 + \mu_3^3\\
                \lambda_3 - \mu_3^3\\
            \end{pmatrix} \in A \subseteq \mathbb{R}^3\\
        \end{align}
    $$
    2. Associativity by associativity of $+$
    3. Neutral Element $e = (0,0,0)$ for $\lambda = 0$, $\mu=0$
    4. Inverse Element $a^{-1} = (-\lambda, -\lambda + (-\mu)^3, -\lambda + (-\mu)^3)$
    5. Commutativity by commutativity of $+$
2. Distributivity by Distributivity of $+$ and $\cdot$
3. Associativity of outer operation by Associativity of $\cdot$
4. Neutral element of outer operation: $1 \in \mathbb{R}$

> (b) $B= \{(\lambda^2, -\lambda^2, 0), \lambda \in \mathbb{R}\}$

This is not a subspace, since $(B, +)$ has no inverse element in $B$ (or in $\mathbb{R}$).
For the inverse element of $(\lambda_1^2, -\lambda_1^2, 0)$, we would need $\lambda_2 = \sqrt{-\lambda_1^2} = \sqrt{-1}\lambda_1 \notin \mathbb{R}$

Similarly, $(B, \cdot)$ is not closed for $-1 \cdot (\lambda^2, -\lambda^2, 0)$.

> (c) Let $\gamma \in \mathbb{R}$. $C= \{(\xi_1, \xi_2, \xi_3), \xi_1 - 2\xi_2 + 3\xi_3 = \gamma \}$

Not a subspace (if $\gamma \neq 0$), since $(C, +)$ is not closed in C:

$(\xi_1, \xi_2, \xi_3) + (\xi_1, \xi_2, \xi_3) = (2\xi_1, 2\xi_2, 2\xi_3)$.

But $2\xi_1 - 4 \xi_2 + 6 \xi_3 = 2 \gamma$.

> (d) $D = \{(\xi_1, \xi_2, \xi_3), \xi_2 \in \mathbb{Z} \}$

Not a subspace, since $D$ is not closed under $\cdot$. $\xi_2$ needs to be in $\mathbb{Z}$, which is not always the case, e.g., multiplying by $\pi$.
