# 2.7 SOLE with Condition

> Find all solutions of the equation system $\boldsymbol{A}\boldsymbol{x} = 12 \boldsymbol{x}$, where:
> $$
    \begin{align}
    \boldsymbol{A} &=
      \begin{bmatrix}
          6 & 4 & 3\\
          6 & 0 & 9\\
          0 & 8 & 0\\
      \end{bmatrix},
      &
      \boldsymbol{x} &= \begin{bmatrix}x_1\\x_2\\x_3\\\end{bmatrix} \in \mathbb{R}^3,
      &
      \sum_{i=1}^3 x_i &= 1
    \end{align}
$$

$$
    \begin{align}
        &\boldsymbol{A} \boldsymbol{x} = 12 \boldsymbol{x}\\
        \iff&
        \begin{bmatrix}
            -6 & 4 & 3\\
            6 & -12 & 9\\
            0 & 8 & -12\\
            1 & 1 & 1\\
        \end{bmatrix}
        \boldsymbol{x} &=
        \begin{bmatrix}
            0\\0\\0\\1
        \end{bmatrix}\\
        \leadsto &
        \left[
            \begin{array}{ccc|c}
            1 & 0 & 0 & 3/8\\
            0 & 1 & 0 & 3/8\\
            0 & 0 & 1 & 2/8\\
            0 & 0 & 0 & 0\\
            \end{array}
        \right]
    \end{align}
$$

```python
import numpy as np

A = np.array([[6, 4, 3], [6, 0, 9], [0, 8, 0]])
x = np.array([[3/8], [3/8], [2/8]])

A @ x
```

```python
12 * x
```
